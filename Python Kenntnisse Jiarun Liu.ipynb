### 4: Selbst Beurteilung: 

**Ich habe das Radardiagramm verwendet, um meine Python-Fähigkeiten anhand von fünf Dimensionen zu bewerten: „Angewandte Kompetenz“, „Grundwissen“, „Kenntnis der neuen Bibliothek“, „Lernfähigkeit“, „Programmierstandardisierung“.** 

```python
# Insert lib
import numpy as np
import matplotlib.pyplot as plt

# Minus Setting
plt.rcParams['axes.unicode_minus'] = False

# ggplot
plt.style.use('ggplot')

# Data Constructing
values = [7.0,7.5,7.0,8.5,6.0]
feature = ['Angewandte Kompetenz','Grundwissen','Kenntnis der neuen Bibliothek','Lernfähigkeit','Programmierstandardisierung']

# Radar Drawing
N = len(values)
angles=np.linspace(0, 2*np.pi, N, endpoint=False)
values=np.concatenate((values,[values[0]]))
angles=np.concatenate((angles,[angles[0]]))
fig=plt.figure()
ax = fig.add_subplot(111, polar=True)
ax.plot(angles, values, 'o-', linewidth=2)
ax.fill(angles, values, alpha=0.20)
ax.set_thetagrids(angles * 180/np.pi, feature)
ax.set_ylim(0,10)
plt.title('Python-Kenntnisse')
ax.grid(True)
plt.show()
```
